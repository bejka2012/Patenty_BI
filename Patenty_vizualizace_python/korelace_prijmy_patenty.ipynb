```python
import pandas as pd
import matplotlib.pyplot as plt

# Naten dat
applicants_file = r'C:\Users\beata\OneDrive\Dokumenty\IT\Digitln akademie Czechitas\Projekt_DA_patenty\Patenty_BI\Patenty_vizualizace_python\applicants_2.xlsx'  # Nahrate skutenou cestu k souboru
licencni_prijmy_file = r'C:\Users\beata\OneDrive\Dokumenty\IT\Digitln akademie Czechitas\Projekt_DA_patenty\Patenty_BI\Patenty_vizualizace_python\licencni_prijmy_inc_kateg.csv'  # Nahrate skutenou cestu k souboru

applicants_data = pd.read_excel(applicants_file)
licencni_prijmy_data = pd.read_csv(licencni_prijmy_file)

# Poet patent na univerzitu
patents_per_university = applicants_data.groupby('ID_univerzita').size().reset_index(name='Pocet_patentu')

# Spojen dataset podle ID univerzity
merged_data = pd.merge(licencni_prijmy_data, patents_per_university, on='ID_univerzita', how='left')

# Korelace mezi potem patent a licennmi pjmy
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Pocet_patentu'], merged_data['Licencni_prijem'], alpha=0.7, color='green')
plt.title('Korelace mezi potem patent a licennmi pjmy')
plt.xlabel('Poet patent')
plt.ylabel('Licenn pjmy')
plt.grid(True)
plt.show()

# Statistick korelace (voliteln)
correlation = merged_data[['Pocet_patentu', 'Licencni_prijem']].corr()
print("Korelace mezi potem patent a licennmi pjmy:\n", correlation)

```
