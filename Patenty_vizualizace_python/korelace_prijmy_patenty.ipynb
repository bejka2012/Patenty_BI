```python
import pandas as pd
import matplotlib.pyplot as plt

# Načtení dat
applicants_file = r'C:\Users\beata\OneDrive\Dokumenty\IT\Digitální akademie Czechitas\Projekt_DA_patenty\Patenty_BI\Patenty_vizualizace_python\applicants_2.xlsx'  # Nahraďte skutečnou cestu k souboru
licencni_prijmy_file = r'C:\Users\beata\OneDrive\Dokumenty\IT\Digitální akademie Czechitas\Projekt_DA_patenty\Patenty_BI\Patenty_vizualizace_python\licencni_prijmy_inc_kateg.csv'  # Nahraďte skutečnou cestu k souboru

applicants_data = pd.read_excel(applicants_file)
licencni_prijmy_data = pd.read_csv(licencni_prijmy_file)

# Počet patentů na univerzitu
patents_per_university = applicants_data.groupby('ID_univerzita').size().reset_index(name='Pocet_patentu')

# Spojení datasetů podle ID univerzity
merged_data = pd.merge(licencni_prijmy_data, patents_per_university, on='ID_univerzita', how='left')

# Korelace mezi počtem patentů a licenčními příjmy
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Pocet_patentu'], merged_data['Licencni_prijem'], alpha=0.7, color='green')
plt.title('Korelace mezi počtem patentů a licenčními příjmy')
plt.xlabel('Počet patentů')
plt.ylabel('Licenční příjmy')
plt.grid(True)
plt.show()

# Statistická korelace (volitelné)
correlation = merged_data[['Pocet_patentu', 'Licencni_prijem']].corr()
print("Korelace mezi počtem patentů a licenčními příjmy:\n", correlation)

```
